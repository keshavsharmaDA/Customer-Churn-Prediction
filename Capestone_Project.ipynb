```python
# Import necessary libraries
import pandas as pd
import numpy as np

employee_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005"],
    "Name": ["John Alter", "Alice Luxumberg", "Tom Sabestine", "Nina Adgra", "Amy Johny"],
    "Gender": ["M", "F", "M", "F", "F"],
    "City": ["Paris", "London", "Berlin", "Newyork", "Madrid"],
    "Age": [25, 27, 29, 31, 30]
}

df_employee = pd.DataFrame(employee_data)

df_employee.to_csv("employee.csv", index=False)


seniority_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005"],
    "Designation Level": [2, 2, 3, 2, 3]
}

df_seniority = pd.DataFrame(seniority_data)


df_seniority.to_csv("seniority.csv", index=False)

# Project DataFrame

project_data = {
    "ID": [
        "A001", "A002", "A003", "A004", "A005", "A002", "A005",
        "A003", "A001", "A003", "A001", "A004", "A004", "A005"
    ],
    "Project": [
        "Project 1", "Project 2", "Project 3", "Project 4", "Project 5",
        "Project 6", "Project 7", "Project 8", "Project 9", "Project 10",
        "Project 11", "Project 12", "Project 13", "Project 14"
    ],
    "Cost": [
        1002000, 2000000, 4500000, 5500000, np.nan,
        680000, 400000, 350000, np.nan, 300000,
        2000000, 1000000, 3000000, 200000
    ],
    "Status": [
        "Finished", "Ongoing", "Finished", "Ongoing", "Finished",
        "Failed", "Finished", "Failed", "Ongoing", "Finished",
        "Failed", "Ongoing", "Finished", "Finished"
    ]
}

df_project = pd.DataFrame(project_data)

df_project.to_csv("project.csv", index=False)
```

```python
# Task 2: Replace Missing Project Costs using Average with For Loop

df_project = pd.read_csv("project.csv")

total = 0
count = 0

for i in range(len(df_project)):
    if pd.isna(df_project.loc[i, 'Cost']):
        if count != 0:
            avg = total / count
        else:
            avg = 0
        df_project.loc[i, 'Cost'] = avg
    else:
        total += df_project.loc[i, 'Cost']
        count += 1

df_project.to_csv("project.csv", index=False)

df_project
```

```python
# Task 3: Split Name into First and Last Name

df_employee = pd.read_csv("employee.csv")

df_employee[['First Name', 'Last Name']] = df_employee['Name'].str.split(' ', n=1, expand=True)

df_employee.drop('Name', axis=1, inplace=True)

df_employee.to_csv("employee.csv", index=False)

df_employee
```

```python
# Task 4: Join all three DataFrames into a single Final DataFrame

df_employee = pd.read_csv("employee.csv")
df_seniority = pd.read_csv("seniority.csv")
df_project = pd.read_csv("project.csv")

df_emp_senior = pd.merge(df_employee, df_seniority, on='ID')

df_final = pd.merge(df_emp_senior, df_project, on='ID')

df_final.to_csv("final.csv", index=False)

df_final
```

```python
# Task 5: Add 5% bonus for Finished projects only

df_final = pd.read_csv("final.csv")

df_final['Bonus'] = df_final.apply(
    lambda row: 0.05 * row['Cost'] if row['Status'].lower() == 'finished' else 0,
    axis=1
)

df_final.to_csv("final.csv", index=False)

df_final
```

```python
# Task 6: Demote failed project heads and remove ineligible ones

df_final = pd.read_csv("final.csv")

df_final['Original Level'] = df_final['Designation Level']

failed_ids = df_final[df_final['Status'].str.lower() == 'failed']['ID'].unique()

df_final['Designation Level'] = df_final.apply(
    lambda row: row['Designation Level'] + 1 if row['ID'] in failed_ids else row['Designation Level'],
    axis=1
)

df_final = df_final[df_final['Designation Level'] <= 4]

df_final.to_csv("final.csv", index=False)

df_final
```

```python
# Task 7: Add prefix to First Name based on Gender (if column exists)

df_final = pd.read_csv("final.csv")

if 'Gender' in df_final.columns:
    df_final['First Name'] = df_final.apply(
        lambda row: "Mr. " + row['First Name'] if row['Gender'] == 'M' else "Mrs. " + row['First Name'],
        axis=1
    )
    
    df_final.drop('Gender', axis=1, inplace=True)
else:
    print("Column 'Gender' not found. Skipping prefixing step.")

df_final.to_csv("final.csv", index=False)

df_final
```

```python
# Task 8: Promote employees whose Age > 29 by decreasing their Designation Level by 1

df_final = pd.read_csv("final.csv")

df_final['Designation Level'] = df_final.apply(
    lambda row: max(1, row['Designation Level'] - 1) if row['Age'] > 29 else row['Designation Level'],
    axis=1
)

df_final.to_csv("final.csv", index=False)

df_final
```

```python
# Task 9: Calculate total project cost for each employee

df_final = pd.read_csv("final.csv")

df_total_cost = df_final.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()

df_total_cost.rename(columns={'Cost': 'Total Cost'}, inplace=True)

df_total_cost.to_csv("total_project_cost.csv", index=False)

df_total_cost
```
